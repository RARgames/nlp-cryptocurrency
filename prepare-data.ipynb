```python
import os
import pandas as pd

def prepare_data(path):
    files = os.listdir(path)
    df = pd.DataFrame()    
    for file in files:
        dfTemp = pd.read_csv(os.path.join(path, file))
        dfTemp.dropna(subset = ["direction"], inplace=True)
        if dfTemp["title"].isnull().sum() > 0:
            dfTemp["title"] = dfTemp["text"]
        dfTemp = dfTemp[['date', 'title', 'direction']]
        df = pd.concat([df, dfTemp], ignore_index=True)
    df.to_csv("data/final.csv")
    return df
```

```python
prepare_data('./data/')
```
